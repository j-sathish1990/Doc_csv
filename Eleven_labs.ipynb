```python
audio = generate(
    text = "Artificial intelligence and machine learning. \
    What's the difference? Are they the same? Well, some people kind of frame the question this way. \
    It's (Artificial Intelligence) AI versus machine learning. Is that the right way to think of this? Or is it AI equals ML? Or \
    is it AI is somehow something different than ML? ", 
    voice = "Matthew",
    )
```

```python
play(audio, notebook = True)
```

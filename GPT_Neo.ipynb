# importing library

```python
from transformers import pipeline # gives required pipeline from huggingface
```

```python
generator = pipeline('text-generation', model ='EleutherAI/gpt-neo-2.7B') # architecture resembles GPT-3
```

```python
generator
```

# generate text using a prompt

```python
prompt = "Please tell me a quote by Isaac Newton "
```

Response variable is set to generator, with key word arguments (max_length : 100 words, do_sample allows sampling within the model) and prompt

```python
response = generator(prompt, max_length  = 100, do_sample = True, temperature= 0.7)
```

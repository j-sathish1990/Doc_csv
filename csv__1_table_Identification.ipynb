```python
start_col = ord(starting_cell[0])- ord('A') + 1
start_row = int(starting_cell[1:])
end_col = ord(end_cell[0]) - ord('A') + 1
end_row = int(end_cell[1:])

```

```python
from prettytable import PrettyTable
table = PrettyTable()
```

```python
for row in sheet.iter_rows(start_row, end_row, start_col, end_col):
  row_values = [cell.value for cell in row]
  if table.field_names == []:
    table.field_names = row_values
  else:
    table.add_row(row_values)
table
```
